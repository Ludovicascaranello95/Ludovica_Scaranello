After closely looking at the data frame, it is clear that there is a large volume of missing values. Some variables are almost empty, resulting in unreliable comparisons and unclear results. This is partly due to the large disparsity from the gathering of data. Most of the variables contain observations that are not drawn in the same year. Another explanation for the oddly large number of missing values would be the failure of the government to report the statistics, or because the information is not available. Other times, missing values are caused by improper collection of data or mistakes during the entry of the data. 

The strategy that is going to be followed for missing values is the imputation by the mean. This means that after being located and flagged, missing values are stored as new variables in the end of the data frame. Furthermore, all the missing values in the data frame are going to be replaced with the value of the mean of the column they belong to. Before replacing the missing values, data needs to be visualized with the help of the histogram. The histogram is an efficient way to visualize the distribution of data and observe its skewness. According to a positive or negative value of skewness, we decide to impute with the mean, rather than the median.  

```python
# The following is a loop that checks every column and detects and flags all missing values. 
# The missing values are then stored into a new column in the end of the dataframe as integers. 
# The cells that are filled with '1' are the flags. 

for col in sub_africa:

    if sub_africa[col].isnull().astype(int).sum() > 0:
        sub_africa['m_'+col] = sub_africa[col].isnull().astype(int)
        
sub_africa.head()
```

```python
# A new column with the sum of missing values per observation is created and stored in the end of the 
# data frame. 
sub_africa['mv_sum'] = sub_africa.isnull().sum(axis = 1)


# checking results
print(f"""

Number of Missing Values per Observation 
----------------------------------------------
{(sub_africa['mv_sum'].value_counts(normalize = True,
                                    sort      = True,
                                    ascending = True)*100).round(2)}
""")


```

```python
# histogram for the column 'Contributing family workers, female'
sns.distplot(a     = sub_africa['Contributing family workers, female (% of female employment)'],
             bins  = 30,
             hist  = True,
             kde   = False,
             rug   = False,
             color = 'black')
# this adds a title
plt.title(label = "Contributing family workers, female (% of female employment)")
# this adds an x-label
plt.xlabel(xlabel = '% of female employment')
# this add a y-label
plt.ylabel(ylabel = 'Frequency')

#Adding the lines for mean and median, as well as legends 
plt.axvline(x = sub_africa['Contributing family workers, female (% of female employment)'].mean(),
            color = 'red')

plt.axvline(x = sub_africa['Contributing family workers, female (% of female employment)'].median(),
            color = 'blue')

plt.legend(labels =  ['mean', 'median'])

# these compile and display the plot so that it is formatted as expected
plt.tight_layout()
plt.show()

```

```python
# histogram for Contributing family workers, total (% of total employment)
sns.distplot(a     = sub_africa['Contributing family workers, total (% of total employment)'],
             bins  = 30,
             hist  = True,
             kde   = False,
             rug   = False,
             color = 'black')
# this adds a title
plt.title(label = "Contributing family workers, total (% of total employment)")
# this adds an x-label
plt.xlabel(xlabel = '% of total employment')
# this add a y-label
plt.ylabel(ylabel = 'Frequency')

#Adding the lines for mean and median, as well as legends 
plt.axvline(x = sub_africa['Contributing family workers, total (% of total employment)'].mean(),
            color = 'red')

plt.axvline(x = sub_africa['Contributing family workers, total (% of total employment)'].median(),
            color = 'blue')

plt.legend(labels =  ['mean', 'median'])


# these compile and display the plot so that it is formatted as expected
plt.tight_layout()
```

```python
# Creating the variables sa1_mean, sa2_mean and sa3_mean. 

sa1_mean = sub_africa['Contributing family workers, female (% of female employment)'].mean()
sa2_mean = sub_africa['Contributing family workers, male (% of male employment)'].mean()
sa3_mean = sub_africa['Contributing family workers, total (% of total employment)'].mean()

# filling NAs with MEAN
sub_africa['Contributing family workers, female (% of female employment)'].fillna(value = sa1_mean,
                           inplace = True)
sub_africa['Contributing family workers, male (% of male employment)'].fillna(value = sa2_mean,
                            inplace = True)
sub_africa['Contributing family workers, total (% of total employment)'].fillna(value = sa3_mean,
                                                                               inplace = True)

# checking to make sure NAs are filled in
print(sub_africa['Contributing family workers, female (% of female employment)'].isnull().any())
print(sub_africa['Contributing family workers, male (% of male employment)'].isnull().any())
print(sub_africa['Contributing family workers, total (% of total employment)'].isnull().any())
```

**Improved sanitation facilities (% of population with access):**
The average for the world is 71.61%, while only 37.50 percent have access in Sub-Saharan countries. According to McKinsey, from 1990 to 2005 life expectancy grew by more than 2 years. It is commonly agreed among experts that the problem starts system wide.  Deciding whether the approach to tackle this problem should be about pursuing “vertical” programs targeted against strong disease outcomes or “horizontal” efforts to strengthen health systems. **Another feature that stands out in this area is the tuberculosis death rate per 10,000 people**. In the Sub-Saharan area, that percentage reaches heights of 36.83%, compared to a 16.3% worldwide average. However, tuberculosis rates are decreasing every year.  It is important to know that in the Sub-Saharan area there are only two types of facilities: dispensaries and health centers. In these two types of facilities, significant shortages of medical supplies and skilled staff as well as absence of electricity and clean water are observed. There was another huge problem with health care workers in this area, specially in Tanzania. The World Health organization estimated that there should be around 92,000 skilled health workers, but the are only 25,400 available. This area does not meet the minimum acceptable physician-to-population ratio of 1-per-5,000. Another studied by the UN calculated that it costs African nations between 21 and 59 thousand dollars to train each doctor. 
***

**Primary completion rate, total (% of relevant age group):**
In some of the poorest regions of the world, children are the most in need of a high-quality education to get ahead, poor learning outcomes often result in higher drop-out rates with large numbers of children not completing school at all, or completing it when more than five years older than the intended graduation age for that level. In sub-Saharan Africa, for example, only 54.79% of primary aged children are enrolled in school at the right grade for their age, when in the rest of the world the average is 86.81%. In sub-Saharan Africa, where education is free, families with a lower disposable income still pay for books and uniforms, and there is the perceived “opportunity cost” of lost income or help with household chores, while children are in school. If parents don’t see a pay-off from their investment in education, children can be pulled out of school before completing a grade, or when transitioning between levels, from say, primary school, to lower secondary.  
***

**Fertility rate, total (births per woman):**
Another feature that makes our area stand out compared to the rest of the world is the fertility rate. The sub - saharan region has an average of  4.95% compared to a 2.94% average throughout the rest of the world.
***

**CO2 emissions (metric tons per capita):**
This data examines the impact of clean energy and inclusive development on CO2 emissions using renewable energy and inequality-adjusted human development index as measures of clean energy and inclusive human development respectively. System generalized method of moments (GMM) was used to analyze data for a sample of 42 sub-Saharan countries.  The result shows that renewable energy has a significant indirect impact on CO2 emissions. This implies that increasing renewable energy usage will dampen CO2 emissions in the region. On the contrary, the result reveals that inclusive human development has a negative and significant impact on CO2 emissions. This implies that important stakeholder such as the marginalized people are not carried along in the campaign for environmental sustainability. The policy implication of the study is as follows.  Firstly, policymakers should subsidize the use of clean energy such as wind, solar, hydroelectric and geothermal to further reduce CO2 emissions. Secondly, the contribution of the less-privileged should be encouraged for inclusive environmental sustainability in sub-Saharan Africa.
